## 写一个小程序来加载数据：运用os来加载和解压和load数据

不用创造datasets folder，系统会自动创建一个foalder 在本文件下。     

```python
import os
import tarfile
from six.moves import urllib
DOWNLOAD_ROOT = "https://raw.githubusercontent.com/ageron/handson-ml/master/"
HOUSING_PATH = "datasets/housing"
HOUSING_URL = DOWNLOAD_ROOT + HOUSING_PATH + "/housing.tgz" 
#在housing文件里下载download root的文件，并命名为housing.tgz的压缩文件

#建立一个解压函数， 这样只需要run一次fetch_housing_data()加载数据，以后要都不用run了
def fetch_housing_data(housing_url=HOUSING_URL, housing_path=HOUSING_PATH):
    if not os.path.isdir(housing_path):
        os.makedirs(housing_path)
        tgz_path = os.path.join(housing_path, "housing.tgz")
    urllib.request.urlretrieve(housing_url, tgz_path)
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=housing_path)
    housing_tgz.close()
# fetch_housing_data()
```

```python
# 使用 Pandas 加载数据。还是用一个小函数来加载数据
import pandas as pd
def load_housing_data(housing_path=HOUSING_PATH):
    csv_path = os.path.join(housing_path,"housing.csv")
    return pd.read_csv(csv_path)
housing = load_housing_data()
```

## 创建测试集/采样（testing set 20%）        
两种采样并且同时创造测试集的方法： 采样和划分训练集和测试集同时发生，且发生在数据预处理之前。     
* 1. 用train_test_split做随机采样 （大样本）
* 2. 用StratifiedShuffleSplit做分层采样 （小样本）。跟据关键属性（这里是新创建的income_cat作为关键属性）做分层抽样。         

训练集，测试集， 验证集 用法：
https://zhuanlan.zhihu.com/p/48976706
* 泛化误差（Generilization Error):把训练好的模型带入实际情况时的误差。
* 怎么判断模型好坏，然后决定是否继续优化：   

    a. 把模型训练好，带入实际情况看泛化误差，如果误差大，就继续训练模型，来回操作。这样成本大，不好。      
    b. 在训练模型的时候，找到一个可以评估模型好坏的measurement。不好，因为训练模型的数据可能有无关属性或噪声干扰or数据没代表性or数据集太少。所以不准确。 
    c. 最好的方法是把数据分成train set和test set，用test set来代替真实环境。把训练好的模型在测试集上训练误差就好。我们默认这个误差为泛化误差的近似。   
    d. 划分测试集和训练集的几点注意：
        i. 在数据预处理（去除缺失值，norm stand）之前进行划分。
        ii. 预处理只针对训练集算参数，然后直接把算好的参数带入测试集。比如，用训练集算出的中位数，来填充训练集和测试集的缺失值。
        iii. 如果想对比多个模型，可以通过对比测试集的generalization error的大小来决定哪个模型更好。

验证集的用法（validation set）：     
* 上面两个模型选一个，可以通过对比test set上的generalization error来选。那么对模型本身进行选择，即选择模型上的超参数，就需要用validation set来选。
* 例子： 假如我们有两个模型，线性模型和神经网络模型，我们知道神经网络的泛化能力要比线性模型要强，我们选择了神经网络模型，但是神经网络中还有很多的需要人工进行选择的参数，比如神经网络的层数和每层神经网络的神经元个数以及正则化的一些参数等等，我们将这些参数称为超参数。这些参数不同选择对模型最终的效果也很重要，我们在开发模型的时候总是需要调节这些超参数。
* 不能用泛化误差的原因： 们最终的目的就是选择泛化能力强的模型，那么我们可以直接通过模型在测试集上的误差来调节这些参数不就可以了。可能模型在测试集上的误差为0，但是你拿着这样的模型去部署到真实场景中去使用的话，效果可能会非常差。们最终的目的就是选择泛化能力强的模型，那么我们可以直接通过模型在测试集上的误差来调节这些参数不就可以了。可能模型在测试集上的误差为0，但是你拿着这样的模型去部署到真实场景中去使用的话，效果可能会非常差。这一现象叫做信息泄露（data leakage)。我们使用测试集作为泛化误差的近似，所以不到最后是不能将测试集的信息泄露出去的. 所以**泛化误差只能比较几个模型之间好坏，不能比较本模型好坏** 。 

总结： 
**也就是说我们将数据划分训练集、验证集和测试集。在训练集上训练模型，在验证集上评估模型，一旦找到的最佳的参数，就在测试集上最后测试一次，测试集上的误差作为泛化误差的近似。**   

三个集的划分方法： 如果当数据量不是很大的时候（万级别以下）的时候将训练集、验证集以及测试集划分为6：2：2；若是数据很大，可以将训练集、验证集、测试集比例调整为98：1：1；但是当可用的数据很少的情况下也可以使用一些高级的方法，比如留出方，K折交叉验证等

什么时候用stratified sampling：
* 1. 可能数据中包含大量的正例样本，仅仅包含10%的反例样本，此时的标签分布很不均匀。所以我们需要采用分层采样的方式进行划分数据集，也就是说保证训练集中既包含一定比例的正例样本又要包含一定比例的负例样本。
* 2. 如本题，有多个分类，每个分类都要抽取一些。

straitfied sampling的原理：
1. 由于上面的情况都只是考虑纯随机采样，如果样本很大，则能表现良好，如果样本比较少，则会有采样偏差的风险。比如要对1000个人做问卷调查，社会人口男女比例为51.3%和48.7%，则采样人数按照这个比例，则应该为513和487，这就是分层采样。如果纯随机采样（即上述按行号，识别码的随机采样），则有12%的可能测试集中女性少于49%或男性多于54%。这样的话就会产生采样偏差。     

    所以sklearn提供了另一个函数StratifiedShuffleSplit（分层随机抽样）这样看来好像确实比刚刚想的纯随机抽样科学很多。
    
    即根据median incomde的分类列income cat分层随机抽样（好像上面的根据男女抽样一样）。
    
    用median incomde是因为专家说median income是对房价很重要的一个属性。**所以我们希望对于总体的抽样，按每一层meidan salary抽取。**
    
    
2. 把上图的连续属性median income变成分类属性，在进行分层采样（ strafied sampling）
3. 创建一个列叫income cat 是median incomde的分类种类 1，2，3，4，5

**方法2： stratfied sampling**
分成采样steps: 
* 把中位数除以1.5（以限制分类的数量，使之变少）
* 创建一个收入分类的属性： 
    用np.ceil(把树变成最小整数，即变成离散量，便于分类）
    用where把所有大于5的归类于5
* 用StraitfiedShuffleSplit:根据收入分类标签房价，并进行分层采样

        a. n_splits是分几组train和test，默认为10.
        b.   参数test_size和train_size是用来设置train/test对中train和test所占的比例.用于分训练集和测试集
        
        c. 用straitfiedshufflesplit做一个函数split，他和数据无关，不用带入数据
        d. split的属性：split.split(data, 根据分层的列）
      
* 最后由于income_cat属性只是我们用于划分的特征，对训练没有任何作用，所以最后需要将加入的income_cat删除   

```python
housing['income_cat']=np.ceil(housing['median_income']/1.5)
```

```python
#dataframe. where(判断式，判断式为false时的替换值，inplace=True是在原数据上做改变) 和if else相反
housing['income_cat'].where(housing['median_income']<5, 5.0, inplace=True)
```

```python
housing['income_cat'].value_counts()/len(housing) #完成把连续属性变成分类属性
```

```python
from sklearn.model_selection import StratifiedShuffleSplit #分大小写
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)

for train_index, test_index in split.split(housing, housing["income_cat"]):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
#完成根据分类属性，把数据采样成train和test set

#之后做ml，就会用strat_train/test_set 来fit
```

```python
strat_train_set['median_house_value'].hist() #和之前整体的hist图分布差不多，体现了分层抽样的重要性
```

```python
#把strat_train/test_set 里的income_cat列删除
for set in (strat_train_set, strat_test_set):
    set.drop(['income_cat'], axis=1, inplace=True)
```

* 4. 特征缩放 

        a. 需要做特征缩放的原因：输入属性量度不同时，会影响ML的性能。这个规律也适用于房产数据：总房间数分布范围是 6 到 39320，而收入中位数只分布在 0 到 15。注意通常情况下我们**不需要对目标值进行缩放**。 
        
        b. 方法1:线性函数归一化（Min-Max scaling/normalization）。
                1. 定义：值被转变、重新缩放，直到范围变成 0 到 1。我们通过减去最小值，然后再除以最大值与最小值的差值，来进行归一化。
                2. 操作：Scikit-Learn 提供了一个转换器MinMaxScaler来实现这个功能。它有一个超参数feature_range，可以让你改变范围，如果不希望范围是 0 到 1。
        c. 方法2:标准化（standardization） 
            
                1. 定义：首先减去平均值（所以标准化值的平均值总是 0），然后除以方差，使得到的分布具有单位方差。
                2. 操作： Scikit-Learn 提供了一个转换器StandardScaler来进行标准化。
                3. 与normalization差别：缺点：与归一化不同，标准化不会限定值到某个特定的范围，这对某些算法可能构成问题（比如，神经网络常需要输入值得范围是 0 到 1）。优点：但是，**标准化受到异常值的影响很小**。例如，假设一个街区的收入中位数由于某种错误变成了 100，归一化会将其它范围是 0 到 15 的值变为 0-0.15，但是标准化不会受什么影响。

```python
#独热编码5:自定义转换器。不懂
from sklearn.base import BaseEstimator, TransformerMixin
rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True): # no *args or **kargs
        self.add_bedrooms_per_room = add_bedrooms_per_room
    def fit(self, X, y=None):
        return self  # nothing else to do
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        population_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, population_per_household,
                         bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, population_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
housing_extra_attribs
```

```python
#sklearn里没有吧dataframe转换成numpy array格式的包，所以我们要自己写一个，
#自己写一个包的过程，就是自定义转换器，写自定义转换器，要用BaseEstimator, TransformerMixin。
#并且写三个方法（功能）：fit()（返回self），transform()，和fit_transform()。
#他是用来获取数值属性的，并把他变成array格式。便于其他估计器使用。下面的pipline有用到。
from sklearn.base import BaseEstimator, TransformerMixin
#输入 属性列名，输出 对应的values
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
```

```python
# 转换流水线
housing_num = housing.drop("ocean_proximity", axis=1)
housing_cat = housing["ocean_proximity"]
from sklearn.pipeline import FeatureUnion
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
```

交叉检验原理：     
* 把训练集分成十份
* 九份为训练集，用来训练模型。 一份为验证集，用来做评估。
* 格式： cross_val_score(模型，train的属性，train的label，scoring=‘评估方法‘）
* 评估： Scikit-Learn 交叉验证功能期望的是效用函数（越大越好）而不是损失函数（越低越好），因此得分函数实际上与 MSE 相反（即负值），这就是为什么前面的代码在计算平方根之前先计算-scores。
* 交叉检验，针对回归/聚类/分类所用的参数不同：回归最常用neg_mean_squared_error. 分类常用accuracy。下表是所有的scoring=参数

```python
def display_scores(scores):
    print('scores:', scores)
    print('mean:',scores.mean())
    print('sd:', scores.std())
display_scores(tree_rmse_scores)
#看起来就不好了，还不如前面的线性回归。 
```

注意到交叉验证不仅可以让你得到模型性能的评估，还能测量评估的准确性（即，它的标准差）。决策树的评分大约是 71200，通常波动有±2361。如果只有一个验证集，就得不到这些信息。但是交叉验证的代价是训练了模型多次，不可能总是这样。

```python
#网格搜索
from sklearn.model_selection import GridSearchCV
param_grid = [
     {'n_estimators': [3, 10, 30], 'max_features': [2, 4, 6, 8]},
    {'bootstrap': [False], 'n_estimators': [3, 10], 'max_features': [2, 3, 4]},
]

forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(housing_prepared, housing_labels)
```

param_grid告诉 Scikit-Learn 首先评估所有的列在第一个dict中的n_estimators和max_features的3 × 4 = 12种组合（不用担心这些超参数的含义，会在第 7 章中解释）。然后尝试第二个dict中超参数的2 × 3 = 6种组合，这次会将超参数bootstrap设为False而不是True（后者是该超参数的默认值）。

总之，网格搜索会探索12 + 6 = 18种RandomForestRegressor的超参数组合，会训练每个模型五次（因为用的是五折交叉验证）。换句话说，训练总共有18 × 5 = 90轮！K 折将要花费大量时间，完成后，你就能获得参数的最佳组合，
